```python
from selenium.webdriver.common.by import By
from time import sleep

# Scroll through the element with the specific ID ('center_col') incrementally
def scroll_and_paginate():
    try:
        # Find the scrollable element by its ID
        scrolling_cube = driver.find_element(By.XPATH, '//*[@id="QA0Szd"]/div/div/div[1]/div[2]/div/div[1]/div/div/div[1]/div[1]')

        # Scroll incrementally
        for _ in range(20):  # You can adjust the range based on how much you want to scroll
            driver.execute_script("arguments[0].scrollTop += 500", scrolling_cube)  # Scroll down 100px
            sleep(0.2)  # Wait for content to load

    except Exception as e:
        print(f"Error during scrolling: {e}")

# Call the function to start scrolling and pagination
scroll_and_extract()

# Lists to store extracted data
names = []
addresses = []
numbers = []

# Lists to store extracted data
names = []
addresses = []
numbers = []

def data_extraction():
    """
    Extracts data (names, addresses, and phone numbers) from the current page of Google Maps search results.
    """
    page_source = BeautifulSoup(driver.page_source, 'html.parser')
    results = page_source.find_all('div', {'class': 'bfdHYd Ppzolf OFBs3e'})
    
    for result in results:
        try:
            # Extract Name
            person_title = result.find('div', class_='NrDZNb')
            name = person_title.get_text() if person_title else 'Null'
            print(f"Name: {name}")
            names.append(name)
        except Exception as e:
            print(f"Error extracting name: {e}")
            names.append('Null')
 
        try:
            # Find all elements with class 'W4Efsd'
            address_elements = result.find_all('div', class_='W4Efsd')
            
            # Ensure there are enough elements before accessing the third one
            if len(address_elements) >= 3:
                address = address_elements[2].get_text()
            else:
                address = 'Null'
            
            print(f"Third Address: {address}")
            addresses.append(address)
        except Exception as e:
            print(f"Error extracting address: {e}")
            addresses.append('Null')

        try:
            # Extract Number
            number_info = result.find('span', class_='UsdlK')
            number = number_info.get_text() if number_info else 'No number found'
            print(f"Number: {number}")
            numbers.append(number)
        except Exception as e:
            print(f"Error extracting number: {e}")
            numbers.append('No number found')

# Scroll and extract data
scroll_and_extract()

# Create a DataFrame and save it to an Excel file
data = {
    "Name": names,
    "Address": addresses,
    "Number": numbers
}
df = pd.DataFrame(data)


```
